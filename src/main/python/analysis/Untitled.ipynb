```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import MultiPolygon
```

```python
kreise = 'C:/Users/david/OneDrive/02_Uni/02_Master/05_Masterarbeit/03_MATSim/01_prep/02_vg250/vg250_neu/VG250_KRS.shp'
```

```python
gdf_kreise = gpd.read_file(kreise)
gdf_kreise['raw'] = gdf_kreise.apply(lambda x:
                                     x['geometry'] if x['geometry'].type == 'MultiPolygon' else
                                     MultiPolygon([x['geometry']]),
                                     axis=1)
```

```python
gdf_kreise = gdf_kreise.to_crs(epsg=25832)
gdf_kreise.columns = gdf_kreise.columns.map(str.lower)
```
