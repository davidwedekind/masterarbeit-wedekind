## Einrichtungsmatrix in Zweirichtungsmatrix überführen

```python
def uni_dir(s):
    
    link = [s["Quelle"], s["Ziel"]]
    link.sort()
    s["Quelle"] = link[0]
    s["Ziel"] = link[1]
    return s
```

```python
df_A_2015 = df_A_2015.apply(uni_dir, axis=1)
df_P_2030 = df_P_2030.apply(uni_dir, axis=1)

df_A_2015.drop_duplicates(subset=["Quelle", "Ziel"], inplace=True)
df_P_2030.drop_duplicates(subset=["Quelle", "Ziel"], inplace=True)

df_A_2015.reset_index(drop=True, inplace=True)
df_P_2030.reset_index(drop=True, inplace=True)
```

```python
df_A_2015.iloc[:,2:len(df_A_2015.columns)] = df_A_2015.iloc[:,2:len(df_A_2015.columns)]*2
df_P_2030.iloc[:,2:len(df_A_2015.columns)] = df_P_2030.iloc[:,2:len(df_A_2015.columns)]*2
```
